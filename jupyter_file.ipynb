```python
img=cv2.imread("invalid.jpg")
img
```

```python
cv2.imshow("Card",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
data=pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
data
```

```python
def validity_check(data):
  exp_lst=[]
  for i in range(len(data['text'])):
    if re.search(r'\d{1,2}/\d{1,4}',data['text'][i]):
      exp_lst.append(data['text'][i])
  if len(exp_lst)==2:
    m1,y1=tuple(exp_lst[0].split("/")) 
    m2,y2=tuple(exp_lst[1].split("/")) 
    if int(y2)>int(y1):
      mon,yr=tuple(exp_lst[1].split("/"))
    else:
      mon,yr=tuple(exp_lst[0].split("/")) 
  else:
    mon,yr=tuple(exp_lst[0].split("/"))    
  curdate=datetime.now()
  curyr=curdate.year
  curmon=curdate.month

  if len(yr)==2:
    curyr=curyr%100
  if int(yr)>curyr:
    print("Card is valid")
  elif int(yr)==curyr:
    if int(mon)>=curmon:
      print("Card is valid")
    else:
      print("Card is invalid")
  else:
    print("Card is invalid")
 


```

```python
validity_check(data)
```
